```python
import seaborn as sns
%matplotlib inline
tip = sns.load_dataset('tips')
```

```python
tip.head(5)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tip,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tip)
```

```python
sns.boxplot(x='day',y='total_bill',data=tip)
```

```python
sns.boxplot(x='day',y='total_bill',data=tip,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tip)
```

```python
sns.violinplot(x='day',y='total_bill',data=tip,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tip,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tip,jitter=True,hue='sex',split=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tip)
sns.swarmplot(x='day',y='total_bill',data=tip,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tip,kind='violin')
```
