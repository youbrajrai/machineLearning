```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('xLabel')
plt.ylabel('yLabel')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OO
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y,'r')
axes1.set_title('Bigger Title')
axes2 = fig.add_axes([0.30,0.31,0.4,0.4])
axes2.plot(y,x,'b')
axes2.set_title('smaller Title')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
fig = plt.figure(figsize=(5,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(5,2))
axes[0].plot(x,y)
axes[1].plot(x,y)
plt.tight_layout()
```

```python
plt.savefig('myPic.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='X axes')
ax.plot(y,x,label='Y-axis')
ax.legend(loc=10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='Purple',linewidth=2,alpha=0.5,linestyle='--',marker='*',markersize=20,markerfacecolor='Yellow',markeredgewidth=4,markeredgecolor='Green')
ax.set_xlim([0,1])
```
